```python
'''
Code To Make The First Version Of Improved Visualization(Pie Chart)

'''

import matplotlib.pyplot as plt


employees = ['1 to 50', '51 to 200','201 to 500','501 to 1000', '1001 to 5000', '5001 to 10000', '10000+']
share = [0.22, 0.14, 0.12, 0.09, 0.15, 0.09, 0.19]

def improved_pie(x_data, y_data):
    plt.figure(figsize=(8, 6), dpi = 300, facecolor='honeydew')
    ax5=plt.subplot(facecolor = 'honeydew')
    dict1 = dict(zip(x_data, y_data))
    #print(dict1)
    dict2 = sorted(dict1.items(),key=lambda x:x[1],reverse=False) 
    #print(dict2)
    new_x = [i[0] for i in dict2 ]
    new_y = [i[1] for i in dict2 ]
    plt.pie(new_y, labels = new_x, autopct='%1.1f%%', 
            shadow=False, startangle=120)
    plt.axis('equal') 

    plt.title(  ' T H E  P R E S E N C E  O F  L E A N \n'
              '----------------------------------------------------\n'
              'by organization size*',
             fontsize = 10
             )

    plt.savefig('pie.jpg')
    plt.show()
    
improved_pie(employees, share)
```

```python
'''
Code To Make The Second Version Of Improved Visualization(Bar Chart)

'''



import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.mlab as mlab 
import numpy as np 


employees = ['1 to 50', '51 to 200','201 to 500','501 to 1000', '1001 to 5000', '5001 to 10000', '10000+']
share = [0.22, 0.14, 0.12, 0.09, 0.15, 0.09, 0.19]


def improved_bar(x_data, y_data):
    plt.figure(figsize=(12, 8), dpi = 300, facecolor='honeydew')
    ax3=plt.subplot(facecolor='honeydew')
    new_y = [100*i for i in y_data]
    plt.xlabel("The amount of employees", fontsize = 15 ) #设置X轴Y轴名称 
    ax3.spines['top'].set_visible(False)
    ax3.spines['left'].set_visible(False)
    ax3.spines['right'].set_visible(False)
    plt.yticks([])


    colors = [
        (0.2, 0.15, 0.7, 0.3),
        (0.2, 0.15, 0.7, 0.4),
        (0.2, 0.15, 0.7, 0.5),
        (0.2, 0.15, 0.7, 0.6),
        (0.2, 0.15, 0.7, 0.7),
        (0.2, 0.15, 0.7, 0.8),
        (0.2, 0.15, 0.7, 0.85)
    ]


    for employee, percentage, changing_color in zip(x_data, new_y, colors):
        ax3.bar(employee , percentage, color = changing_color )
        ax3.yaxis.set_major_formatter(ticker.PercentFormatter())
        plt.text(employee, percentage + 0.1, '%.0f' % percentage + '%', ha='center', va= 'bottom',fontsize=11) 



    plt.text( '51 to 200' , 20, ' T H E  P R E S E N C E  O F  L E A N \n'
              '----------------------------------------------------',
             fontsize = 18
             )
    plt.text('51 to 200', 19.5, 'by organization size (measured by employee numbers) *')

    plt.savefig('improved_bar.jpg')

    plt.show()

    
    
improved_bar(employees, share)
```

```python
import cv2
import numpy as np
#mean hash
def aHash(img):

    img = cv2.resize(img, (8, 8), interpolation=cv2.INTER_CUBIC)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    s = 0
    hash_str = ''
    # get the num of pixels
    for i in range(8):
        for j in range(8):
            s = s + gray[i, j]
    # get mean gray
    avg = s / 64
    # make 0/1 squence
    for i in range(8):
        for j in range(8):
            if gray[i, j] > avg:
                hash_str = hash_str + '1'
            else:
                hash_str = hash_str + '0'
    return hash_str
#Difference hash algorithm
def dHash(img):
    #make the image to 8*8
    img=cv2.resize(img,(9,8),interpolation=cv2.INTER_CUBIC)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    hash_str=''

    for i in range(8):
        for j in range(8):
            if   gray[i,j]>gray[i,j+1]:
                hash_str=hash_str+'1'
            else:
                hash_str=hash_str+'0'
    return hash_str


def cmpHash(hash1,hash2):
    n=0
    # -1 means something goes wrong
    if len(hash1)!=len(hash2):
        return -1
    #judge
    for i in range(len(hash1)):

        if hash1[i]!=hash2[i]:
            n=n+1
    return n

img1=cv2.imread('original.png')
img2=cv2.imread('replication.jpg')
hash1= aHash(img1)
hash2= aHash(img2)
print(hash1)
print(hash2)
n=cmpHash(hash1,hash2)
print ('Mean hash algorithm difference:'+ str(n))

hash1= dHash(img1)
hash2= dHash(img2)
print(hash1)
print(hash2)
n=cmpHash(hash1,hash2)
print ('Difference hash algorithm difference:' + str(n))



```
