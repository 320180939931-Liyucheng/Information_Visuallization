```python
import matplotlib.pyplot as plt
```

```python
employees = ['1 to 50', '51 to 200','201 to 500','501 to 1000', '1001 to 5000', '5001 to 10000', '10000+']
share = [0.22, 0.14, 0.12, 0.09, 0.15, 0.09, 0.19]
```

```python
fig5, ax5 = plt.subplots()
dict1 = dict(zip(employees, share))
#print(dict1)
dict2 = sorted(dict1.items(),key=lambda x:x[1],reverse=False) 
#print(dict2)
new_employees = [i[0] for i in dict2 ]
new_share = [i[1] for i in dict2 ]
ax5.pie(new_share, labels=new_employees, autopct='%1.1f%%', 
        shadow=True, startangle=120)
ax5.axis('equal') 
```
