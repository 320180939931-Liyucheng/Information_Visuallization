```python
def Fibonacci(x):
    if x == 0:
        return 0
    if x == 1:
        return 1
    else:
        result_list = [0,1]
        count = 3
        while count <= (x+1):
            temp = result_list[count - 2] +  result_list[count - 3]
            result_list.append(temp)
            count+=1
    return result_list[x]


for i in range (0,15):
    print(Fibonacci(i))

# Artists_10   define a function that can caclucate F(x)
```

```python
from matplotlib.patches import  Rectangle
fig1, ax1  = plt.subplots()


for i in range(0,15):
    temp = Fibonacci(i)
    rec = Rectangle((-0.5*temp,-0.5*temp),temp , temp, fill = 0)
    ax1.set_xlim((-200,200))
    ax1.set_ylim((-200,200))
    ax1.add_patch(rec)



# Artists_11  
```

```python
fig2, ax2  = plt.subplots()
ax2.set_xlim((-500,500))
ax2.set_ylim((-500,500))

for i in range(0,15):
    temp = Fibonacci(i)
    ax2.add_patch(Rectangle((0, 0), temp, temp, angle = 90*(i+2), fill=False))
# Artists_12
```

```python
fig3, ax3  = plt.subplots()

ax3.set_xlim((-60,60))
ax3.set_ylim((-60,60))
change_x = 0
change_y = 0

for i in range(0, 16):
    temp = Fibonacci(i)
    ax3.add_patch(Rectangle((change_x, change_y), temp, temp, angle = 90*(i+1), fill=False, edgecolor = 'red'))
    if i%4 == 0:
        change_x-=temp
        change_y+=temp
    if i%4 == 1:
        change_x-=temp
        change_y-=temp
    if i%4 == 2:
        change_x+=temp
        change_y-=temp
    if i%4 == 3:
        change_x+=temp
        change_y+=temp

# Artists_13
```

```python
from matplotlib.patches import Wedge
fig4, ax4  = plt.subplots()
ax4.set_xlim((-600,600))
ax4.set_ylim((-600,600))
change_x = 0
change_y = 0

for i in range(0, 18):
    temp = Fibonacci(i)
    gap = Fibonacci(i+1) - temp
    ax4.add_patch(Wedge((change_x, change_y), temp, 90*i, 90*(i+1), fill=True, edgecolor = 'red', facecolor = 'gray'))
    if i%4 == 0:
        change_y -= gap
    if i%4 == 1:
        change_x += gap
    if i%4 == 2:
        change_y += gap
    if i%4 == 3:
        change_x -= gap
        
# Artists_13
```
