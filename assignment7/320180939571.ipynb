```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
money_values = [3000, 5000, 4000, 6000, 8000, 7000, 9000, 5000, 7000, 8000]
#1-2
```

```python
fig, ax = plt.subplots()
ax.plot(months, money_values)
#1-3
```

```python
fig1, ax1 = plt.subplots() 
ax1.plot(months, money_values, color = 'black', marker = "o", markersize=6, markerfacecolor = 'white')
#1-4
```

```python
ax1.grid()
fig1
#1-5
```

```python
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
fig1
#1-6
```

```python
ax1.set_ylim(0,10000)
ax1.set_yticks([1000*i for i in range(11)])
fig1
#1-7
```

```python
new_list = [f"{value:,}" for value in [1000*i for i in range(11)]]
ax1.set_yticklabels([str(i) for i in new_list])
print(new_list)
fig1
#1-8
```

```python
for n, k, m in zip(months, money_values, [f"{value:,}" for value in money_values]):
    ax1.text(n, k, str(m), verticalalignment = 'top')
fig1
#1-9
```

```python
fig2, ax2 = plt.subplots() 
ax2.plot(months, money_values, color = 'black', marker = "o", markersize=6, markerfacecolor = 'white')
ax2.grid()
ax2.grid(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
for n, k, m in zip(months, money_values, [f"{value:,}" for value in money_values]):
    ax2.text(n, k, str(m), verticalalignment = 'top')
# II-1
```

```python
fig3, ax3 = plt.subplots() 
ax3.plot(months, money_values, color = 'black', marker = "o", markersize=6, markerfacecolor = 'white')
ax3.grid()
ax3.grid(False) 
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)
new_list = [f"{value:,}" for value in money_values]
ax3.text(months[6], money_values[6], new_list[6], verticalalignment = 'top')
# II-2
```

```python
fig4, ax4 = plt.subplots() 
ax4.plot(months, money_values, linewidth = '0.5', color = 'black', marker = "o", markersize=6, markerfacecolor = 'white')
ax4.grid()
ax4.grid(False) 
ax4.spines['right'].set_visible(False)
ax4.spines['top'].set_visible(False)
new_list = [f"{value:,}" for value in money_values]
ax4.text(months[6], money_values[6], new_list[6], verticalalignment = 'top')
# II-3
```

```python
fig5, ax5 = plt.subplots() 
ax5.plot(months, money_values, linewidth = '0.5', color = 'black', marker = "o", markersize=10, markerfacecolor = 'white')
ax5.grid()
ax5.grid(False) 
ax5.spines['right'].set_visible(False)
ax5.spines['top'].set_visible(False)
new_list = [f"{value:,}" for value in money_values]
ax5.text(months[6], money_values[6], new_list[6], verticalalignment = 'top')
ax5.spines['left'].set_visible(False)
plt.yticks([])
# II-4
```

```python
fig6, ax6 = plt.subplots() 
ax6.plot(months, money_values, linewidth = '0.5', color = 'black', marker = "o", markersize=10, markerfacecolor = 'white')
ax6.grid()
ax6.grid(False) 
ax6.spines['right'].set_visible(False)
ax6.spines['top'].set_visible(False)
new_list = [f"{value:,}" for value in money_values]
ax6.text(months[6], money_values[6], str(new_list[6])+'(USD $)', verticalalignment = 'top')
ax6.spines['left'].set_visible(False)
plt.yticks([])
# II-5
```
