```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
movie_types = ["SciFi", "Comedy", "Drama", "Action", "Romance"]
movie_preferences = [40, 42, 51, 2, 59]
```

```python
fig, ax = plt.subplots()
ax.plot(movie_types, movie_preferences)
```

```python
fig1, ax1 = plt.subplots()
plt.bar(movie_types, height=movie_preferences, width=0.5)
plt.show()
#Exercise_5
```

```python
total = sum(movie_preferences) 
movie_preferences_percentage = [100*pref/total for pref in movie_preferences] 
print(movie_preferences_percentage)
#Exercise_6
```

```python
fig2, ax2 = plt.subplots()
plt.bar(movie_types, height = (movie_preferences_percentage), width=0.5)
plt.show()
#Exercise_7
```

```python
import matplotlib.ticker as ticker
fig3, ax3 = plt.subplots()
ax3.bar(movie_types, movie_preferences_percentage)
ax3.yaxis.set_major_formatter(ticker.PercentFormatter())
#Exercise_8
```

```python
fig4, ax4 = plt.subplots()
ax4.pie(movie_preferences_percentage, labels=movie_types, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax4.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
#Exercise_9
```

```python
fig5, ax5 = plt.subplots()
dict1 = dict(zip(movie_types, movie_preferences_percentage))
print(dict1)
dict2 = sorted(dict1.items(),key=lambda x:x[1],reverse=False) 
print(dict2)
new_movie_types = [i[0] for i in dict2 ]
new_movie_preferences_percentage = [i[1] for i in dict2 ]
ax5.pie(new_movie_preferences_percentage, labels=new_movie_types, autopct='%1.1f%%', 
        shadow=True, startangle=0)
ax5.axis('equal')  
#Exercise_10
```
